```python
classes_list= ["black_grass", "charlock", "cleavers", "common_chickweed", "common_wheat", "fat_hen", "loose_silky_bent", "maize",
"scentless_mayweed", "shepherds_purse", "small_flowered_cranesbill", "sugar_beet"]
```

```python
#generate 900 number of 3 digits and shuffle them randomly to not get repetitions
ident1 = []
newList=[]
detection11 = []
for i in range(900):
    ident1.append(f'{i:03}')
    random.shuffle(ident1)
    detect= random.choice(classes_list)
    detection11.append(detect)
    
print(ident1)
print("\n")
print(detection11)

myNew = [] 
myDup = []

for x in ident1:
    if x not in myNew:
        myNew.append(x)
    else:
        myDup.append(x)

print("\nthe dupplication we have are ",myDup)
```
