```python
import csv
```

```python
import random
```

```python
classes_list= ["black_grass", "charlock", "cleavers", "common_chickweed", "common_wheat", "fat_hen", "loose_silky_bent", "maize",
"scentless_mayweed", "shepherds_purse", "small_flowered_cranesbill", "sugar_beet"]

```

```python
ident=[]
detection1=[]
detection2=[]
for i in range(900):
    ident.append(f'{i:03}')
    #file.write(f'{i:03}'"\n")
    detect= random.choice(classes_list)
    detectt= random.choice(classes_list)
    detection1.append(detect)
    detection2.append(detectt)

print(ident)
print("\n")
print(detection1)
print("\n")
print(detection2)
```

```python
with open ("test2.csv", "a", newline="") as f:
    writer= csv.writer(f, delimiter =';')
    writer.writerows(zip(ident, detection1))
    
f.close()   
```

```python
with open ("test1.csv", "a", newline="") as f1:
    writer= csv.writer(f1, delimiter=';')
    writer.writerows(zip(ident, detection2))
    
f1.close()
```

```python
with open("test1.csv", "r") as t1:
    test_1_csv = t1.readlines()
with open("test2.csv", "r") as t2:
    test_2_csv = t2.readlines()
    
with open("diff.csv", "w") as outFile:
    line_diff = 0
    line_sim = 0
    for line in test_1_csv:
        if line not in test_2_csv:
            outFile.write(line)
            line_diff += 1
        else:
            line_sim += 1
            
outFile.close()
t1.close()
t2.close()
```
